```python
def changing_files_names(location, food_name):

    i = 0

    for filename in os.listdir(location): 
        dst = food_name + str(i) + ".jpg"
        src = filename 

        # rename() function will 
        # rename all the files 
        os.rename(location+src, location+dst) 
        i += 1
  
```

```python
changing_files_names("../../tortilla/", 'tortilla')
```
